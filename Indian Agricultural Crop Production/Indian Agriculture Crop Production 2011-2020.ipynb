```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
agri=pd.read_csv("E:\MBA Semester II\Python\Data sets\India Agriculture Crop Production 2010-2020.csv")
```

```python
agri
```

```python
agri.info()
```

```python
agri.describe()
```

```python
plt.figure(figsize=(15,20))
SP=agri.groupby('State')['Production'].agg('mean')
SP.sort_values(ascending=True,inplace=True)
SP.plot(kind='barh', color='red')
```

### The above bar plot shows the overall Production in state wise between the year of 2010 to 2020. Punjab is the more production state combare to the all other states.

```python
plt.figure(figsize=(15,10))
CP=agri.groupby('Crop')['Production'].agg('mean')
CP.sort_values(ascending=True,inplace=True)
CP.plot(kind='barh',color='green')
```

### The above bar plot shows the overall Production in Crops.Sugarcane is the more producable crop in all states.

```python
plt.figure(figsize=(15,10))
Sea=agri.groupby('Season')['Production'].agg('mean')
Sea.sort_values(ascending=False,inplace=True)
Sea.plot(kind='bar',color='purple')
```

### The above bar plot shows the overall Production in season. Whole year crops only produced more crops.

```python
plt.figure(figsize=(15,10))
YP=agri.groupby('Year')['Production'].agg('mean')
YP.sort_values(ascending=False,inplace=True)
YP.plot(kind='bar',color='violet')
```

### The above bar plot shows the overall Production in year wise. In 2014-15 period only the more crops are produced combare to the all years. 

### The above bar plot shows the how much farming areas are in states. The most number of farming areas are there in Punjab state only. 

```python
plt.figure(figsize=(15,10))
CY=agri.groupby('Crop')['Yield'].agg('mean')
CY.sort_values(ascending=True,inplace=True)
CY.plot(kind='barh')
```

### The above bar plot shows the how much Yield was get it in the crops. Sugarcane is the more yeild crop when combare to the other crops.
### The yield was calculated in how much the production / the total area.
